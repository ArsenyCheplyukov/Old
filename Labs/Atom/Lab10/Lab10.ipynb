```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

#### **1. T=600K:**

```python
def plot_save_interpolate(x, y, *graph_data):
    graph_name, axis_x, axis_y, save_name = graph_data
    x_array = np.linspace(min(x), max(x), 10000)
    f_interp = interp1d(x, y, bounds_error=True, kind = 'cubic')
    #($10^6$)
    plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
    plt.ylabel(axis_y, fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
    plt.xlabel(axis_x, fontsize=9, style='italic')
    plt.scatter(x, y, c="red", label='Измерянные величины')
    plt.plot(x_array, f_interp(x_array), color="black", label='Сглаженные величины')
    plt.legend()
    plt.title(graph_name, fontsize=12, fontweight='bold', style='italic')
    plt.savefig(save_name+'.jpg', format='jpg', dpi=300)
    plt.show()
    return f_interp
```

```python
y = np.array([0.7, 5.2, 9.4, 16.1, 22.3, 37.4, 43.4, 24.3, 25, 36.6, 58.2, 73.5, 80.5, 69, 66.7, 73, 86.7, 97.3,96.4,84.2,85.2]) / 1000
x = np.array([0, 2.3, 3.9, 6.3, 7.8, 10.2, 11.8, 14.2, 15.7, 18.1, 20.5, 22, 24.4, 26, 28.4, 29.9, 32.3, 33.9, 36.3, 38, 39.9])

x_array = np.linspace(min(x), max(x), 10000)
f_interp = interp1d(x , y, bounds_error=True, kind = 'cubic')

print(f_interp(x_array))
```

```python
#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"${I_а}$, мА", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel(r"U, В", fontsize=9, style='italic')

plt.scatter(x, y, c="red", label='Измерянные величины')
plt.plot(x_array, f_interp(x_array), color="black", label='Сглаженные величины')

# y_full = f_interp(x_array)
# all_extr = np.array([k for k in range(1, len(y_full)-1) if y_full[k-1] < y_full[k] < y_full[k+1]])[]
# for i in all_extr:
#     plt.axhline(y=y_full[i], color='black', linestyle='--')
#     plt.axvline(x=x_array[i], color='black', linestyle='--')
#print((f(10)-f(0))/10, f(0))
plt.legend()
plt.title("Зависисмость тока анода от напряжения\n между катодом и сеткой лампы", fontsize=12, fontweight='bold', style='italic')
plt.savefig('cubic_dependancy.jpg', format='jpg', dpi=300)
plt.show()
```

#### **2. T=600К:**

```python
y = np.array([178, 174, 144, 126, 124, 116, 106, 92, 66, 62, 37, 12, 4, 1, 0, 0, 0]) / 1000
x = np.array([0, 0.8, 2.4, 4, 4.8, 6.9,8.1, 9.3, 11.8, 12.2, 14.2, 16.3, 17.5, 19.9, 20.3, 22.2, 22.8])

x_array = np.linspace(min(x), max(x), 10000)
f_interp = interp1d(x , y, bounds_error=True, kind = 'cubic')

print(f_interp(x_array))

#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"${I_а}$, мА", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel(r"U, В", fontsize=9, style='italic')

plt.scatter(x, y, c="red", label='Измерянные величины')
plt.plot(x_array, f_interp(x_array), color="black", label='Сглаженные величины')

df = f_interp._spline.derivative(nu=1)
plt.title("Зависисмость анодного тока от\n задерживающего потенциала", fontsize=12, fontweight='bold', style='italic')
plt.savefig('cubic_dependancy_3.jpg', format='jpg', dpi=300)
plt.show()

plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"$\dfrac{\Delta I}{\Delta U}$, $\dfrac{мА}{В}$", fontsize=12, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel(r"U, В", fontsize=12, style='italic')

plt.plot(x_array, -df(x_array), color="black")

plt.title("функцию распределения свободных\n электронов по энергиям", fontsize=14, fontweight='bold', style='italic')
plt.tight_layout()
plt.savefig('cubic_dependancy_7.jpg', format='jpg', dpi=300)
plt.show()
```
