```python
y = (np.array([191, 191, 180, 166, 150, 130, 120, 110, 105, 98, 90, 85, 82, 80, 70, 58, 49, 40, 35]))
x = (np.array([0.050, 0.100, 0.150,0.160, 0.170, 0.175, 0.180, 0.185, 0.190, 0.195, 0.200, 0.205, 0.210, 0.215, 0.270, 0.325, 0.380, 0.435, 0.535]))
f = interp1d(x[2:], y[2:], bounds_error=False, kind = 'cubic')
f2 = np.poly1d(np.polyfit(1/x[:3], y[:3], 2))
x, y
```

```python
x_array = np.linspace(min(x), max(x), 10000)
#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"${I_A}$, А", fontsize=14, style='italic')
#plt.ylabel(r"$\dfrac{1}{a_2}$, $м^{-1}$", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel(r"${I_c}$, A", fontsize=14, style='italic')

plt.scatter(x, y, c="black", label='Измерянные величины')
plt.plot(x_array[:2000], f2(x_array)[:2000], color="red", label='Сглаженные величины')
plt.plot(x_array[2000:], 1/f(x_array)[2000:], color="red")

#plt.axhline(y=1.4445, color='black', linestyle='--')
#plt.axvline(x=79, color='black', linestyle='--')
#print((f(10)-f(0))/10, f(0))
plt.legend()
plt.title("Зависимость тока анода\n от тока соленоида", fontsize=14, fontweight='bold', style='italic')
plt.tight_layout()
plt.savefig('unknown_dependancy.jpg', format='jpg', dpi=300)
plt.show()
```
