```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
from matplotlib.pyplot import figure
figure(figsize=(40, 30), dpi=80)
```

```python
y = 1/(np.array([28, 16, 14, 10, 8, 7, 6, 6, 5, 5, 4, 4, 3, 3, 3]))
x = (np.array([0.11, 0.12, 0.13, 0.14, 0.15, 0.16, 0.17, 0.18, 0.19, 0.20, 0.21, 0.22, 0.23, 0.24, 0.25]))
f = np.poly1d(np.polyfit(x, y, 1))
x, y
```

```python
x_array = np.linspace(min(x), max(x), 10000)
#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel(r"$\dfrac{1}{D}$, $мм^{-1}$", fontsize=8, style='italic')
#plt.ylabel(r"$\dfrac{1}{a_2}$, $м^{-1}$", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel("∆y, мм", fontsize=8, style='italic')

plt.scatter(x, y, c="black", label='Измерянные величины')
plt.plot(x_array, f(x_array), color="red", label='Сглаженные величины')

#plt.axhline(y=1.4445, color='black', linestyle='--')
#plt.axvline(x=79, color='black', linestyle='--')
#print((f(10)-f(0))/10, f(0))
plt.legend()
plt.title("Зависисмость обратной ширины первого дифракционный максимума\n от значения ширины щели", fontsize=8, fontweight='bold', style='italic')
plt.savefig('linear_dependancy.jpg', format='jpg', dpi=300)
plt.show()
```

```python
#($10^6$)
plt.grid(color='black', linestyle='--', linewidth=0.1, which='both')
plt.ylabel("D, мм", fontsize=8, style='italic')
#plt.ylabel(r"$\dfrac{1}{a_2}$, $м^{-1}$", fontsize=9, style='italic') #\dfrac{1}{a₁}$, $см^{-1}
plt.xlabel("∆y, мм", fontsize=8, style='italic')

plt.scatter(x, 1/y, c="black", label='Измерянные величины')
plt.plot(x_array, 1/f(x_array), color="red", label='Сглаженные величины')

#plt.axhline(y=1.4445, color='black', linestyle='--')
#plt.axvline(x=79, color='black', linestyle='--')
#print((f(10)-f(0))/10, f(0))
plt.legend()
plt.title("Зависисмость ширины первого дифракционный максимума\n от значения ширины щели", fontsize=8, fontweight='bold', style='italic')
plt.savefig('inverse_dependancy.jpg', format='jpg', dpi=300)
plt.show()
```
